### 4. Limpieza y formateo de datos
Dado que algunos archivos de Excel tienen filas vacías o desorganizadas se necesitara realizar una limpieza y formateo

#### 4.1 Para el archivo 'Level1'

```python
# Verificar las primeras 8 filas para ver donde estan los encabezados
level1.head(8)
```

```python
# Verificar las ultimas 5 filas para ver donde estan los encabezados
level1.tail(5)
```

```python
level1_cleaned = level1.dropna(how='all')  #Eliminar filas completamente vacías
level1_cleaned = pd.read_excel('/content/drive/My Drive/Control3/appendData/appendData/level1.xlsx', skiprows=4) #Se omiten las primero 4 filas
level1_cleaned = level1_cleaned.drop(columns=['Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15'], errors='ignore') #Se omiten columnas vacias/caches
level1_cleaned = level1_cleaned.reset_index(drop=True) #Resetear el índice después de eliminar filas
```

```python
level1_cleaned.head()
```

```python
#Se observa varios decimales por lo que se hara un redondeo a la cantidad de decimales original del archivo

# Verificar los nombres de las columnas en el DataFrame
print(level1_cleaned.columns)
```

```python
# Se define una funcion para redondear a partir de una fila en especifica
def round_from_row(df, column_name, decimals, start_row):
    df.loc[start_row:, column_name] = df.loc[start_row:, column_name].apply(lambda x: round(x, decimals))
    return df

# Aplicar el redondeo solo a partir de la fila 2 en adelante de las columnas necesarias asi como la cantidad de decimales a usar
level1_cleaned = round_from_row(level1_cleaned, 'Life expectancy at birth', 1, 2)
level1_cleaned = round_from_row(level1_cleaned, 'Expected years of schooling', 1, 2)
level1_cleaned = round_from_row(level1_cleaned, 'Mean years of schooling', 1, 2)
level1_cleaned = round_from_row(level1_cleaned, 'Gross national income (GNI) per capita', 3, 2)

# Se visualiza el DataFrame con los valores redondeados a partir de la fila 2
level1_cleaned.head(10)
```

```python
#Finalmente, añadimos una fila adicional para insertar el mensaje VERY HIGH HUMAN DEVELOPMENT ubicado en la columna 'Country'

# Mensaje que se desea añadir para la determinada cantidad de paises
mensaje1 = 'VERY HIGH HUMAN DEVELOPMENT'

# Se crea una nueva fila con todas las celdas vacías (NaN), excepto en la columna donde se quiere insertar el mensaje
nueva_fila1 = pd.DataFrame([[np.nan] * level1_cleaned.shape[1]], columns=level1_cleaned.columns)
nueva_fila1.at[0, 'Country'] = mensaje1

# Insertar la nueva fila en la posición deseada (fila 2)
level1_cleaned = pd.concat([level1_cleaned.iloc[:2], nueva_fila1, level1_cleaned.iloc[2:]]).reset_index(drop=True)

# Ver DataFrame actualizado
level1_cleaned.head(10)
```

### 7. Reporte mensual de fallecidos

```python
# Ya se cargo el archivo anteriormente, se visualiza las primeras filas
fallecidos_df.head()
```

```python
# Se convierte la columna de fecha de fallecimiento a formato "datetime"
fallecidos_df['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_df['FECHA_FALLECIMIENTO'], format='%Y-%m-%d', errors='coerce')

# Se crea nuevas columnas de mes y año
fallecidos_df['mes'] = fallecidos_df['FECHA_FALLECIMIENTO'].dt.month
fallecidos_df['año'] = fallecidos_df['FECHA_FALLECIMIENTO'].dt.year

# Se agrupa por mes y año, y contar los fallecimientos
reporte_mensual = fallecidos_df.groupby(['año', 'mes']).size().reset_index(name='fallecidos')

reporte_mensual.head()
```

```python
# Convertir la columna 'FECHA_FALLECIMIENTO' al formato de fecha
fallecidos_df['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_df['FECHA_FALLECIMIENTO'], format='%Y%m%d')

# Crear una columna 'Mes_Año' para identificar el mes y año del fallecimiento
fallecidos_df['Mes_Año'] = fallecidos_df['FECHA_FALLECIMIENTO'].dt.to_period('M')

# Agrupar por mes y tipo de muerte (columna 'CLASIFICACION_DEF'), contando los fallecidos
monthly_deaths_by_type = fallecidos_df.groupby(['Mes_Año', 'CLASIFICACION_DEF']).size().unstack(fill_value=0)

# Mostrar el reporte (primeras 10 filas)
print(monthly_deaths_by_type.head(10))

# Mostrar el reporte completo
monthly_deaths_by_type
```
